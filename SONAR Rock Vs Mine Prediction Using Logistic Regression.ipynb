In these project we are going to build a system that can predict whether an object is either Rock or Mine with SONAR data.For this case we are using Logistic Regression Model for our prediction.

## Import Libraries

```python
import numpy as np
import pandas as pd
```

## Import Data

```python
data=pd.read_csv("C:\python\Datasets\Sonar Data.csv",header=None)
```

```python
pd.set_option('display.max_columns',None)
data.head()
```

## Data Preprocessing

## Divide the data into Dependent and Indenpendent Variable

```python
x=data.iloc[:,:-1].values
x
```

```python
y=data.iloc[:,-1]
```

```python
y=y.replace({'R':0,'M':1}).values
```

```python
y
```

## Split the data into Training and Testing 

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,stratify=y,random_state=1)
```

```python
x_train,x_test,y_train,y_test
```

## Fit the Logistic Regression model to training data

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

## Prediction  for Train data

```python
y_pred_train=model.predict(x_train)
y_pred_train
```

## Accuracy on Train Data

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
acc=accuracy_score(y_train,y_pred_train)*100
acc
```

```python
print(f'Accuracy score for the training data is {acc} %. Which is quite good.')
```

## Prediction for Test Data

```python
y_pred=model.predict(x_test)
```

```python
np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1)
```

## Accuracy on Test data

```python
acc=accuracy_score(y_test,y_pred)*100
acc
```

```python
print('Accuracy score for the testing data is {} %'.format(acc)) 
```
