In this project we will create a sytem where a user can give the name of his favourite movie and based on this input we are going to recommend certain movies to him. <br>
There are three main types of movie recommendation.
1) Content Based Recommendation System. <br>
2) Popularity Based Recommendation System. <br>
3) Collaborative Recommendation System.

### Here we are going to select only some of the relevent features for building Movie Recommendation System.

```python
# Selecting the relevent features for recommendation system
selected_features=['genres','keywords','tagline','cast','director']
```

```python
selected_features
```

```python
# Replacing the null values with null string
for feature in selected_features:
    data[feature]=data[feature].fillna('')
```

```python
# Combining all the 5 selected features
combined_features=data['genres']+''+data['keywords']+''+data['tagline']+''+data['cast']+''+data['director']
```

```python
combined_features
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Converting the text data to feature vectors
vectorizer=TfidfVectorizer()
```

```python
feature_vectors=vectorizer.fit_transform(combined_features)
print(feature_vectors)
```

## Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Getting the similarity scores using cosine similarity
similarity=cosine_similarity(feature_vectors)
```

```python
similarity
```

```python
similarity.shape
```

```python
# Getting the movie name from the user
movie_name=input('Enter your favourite movie name:')
```

```python
# Creating list with all the movie names given in the dataset
list_of_all_titles=data['title'].tolist()
```

```python
print(list_of_all_titles)
```

```python
import difflib
```

```python
# Finding the close match for the movie name given by the user
find_close_matches=difflib.get_close_matches(movie_name,list_of_all_titles)
```

```python
find_close_matches
```

```python
close_match=find_close_matches[0]
```

```python
close_match
```

```python
# Finding the index of the movie with the title
index_of_the_movie=data[data.title==close_match]['index'].values[0]
```

```python
index_of_the_movie
```

```python
# Getting the list of a similar movies
similarity_score=list(enumerate(similarity[index_of_the_movie]))
```

```python
similarity_score
```

```python
len(similarity_score)
```

```python
# Sorting the movies based on their similarity score  (By assending order)
sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)   # here x represents similarity_score 
```

```python
sorted_similar_movies
```

```python
# Print the name of similar movies based on the index.
print('Movies suggested for you:\n')
i=1
for movie in sorted_similar_movies:
    index=movie[0]
    title_from_index=data[data.index==index]['title'].values[0]
    if (i<30):
        print(i, '-',title_from_index)
        i+=1
```

## Movie Recommendation System

```python
movie_name=input('Enter your favourite movie name:')
list_of_all_titles=data['title'].tolist()
find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles)
close_match=find_close_match[0]
index_of_the_movie=data[data.title==close_match]['index'].values[0]
similarity_score=list(enumerate(similarity[index_of_the_movie]))
sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)   # here x represents similarity_score

# Print the name of similar movies based on the index.
print('Movies suggested for you:\n')
i=1
for movie in sorted_similar_movies:
    index=movie[0]
    title_from_index=data[data.index==index]['title'].values[0]
    if (i<30):
        print(i, '-',title_from_index)
        i+=1
```
